# Pre processing function

```python
# Preprocessing function
def preprocess_train_valid(df):
    # Work on a copy
    df = df.copy() #df = X_train_val.copy()  UPDATED
    
    #drop duplicates
    df = df.drop_duplicates()
    
    #Handling NaN - specific columns
    median_pctemployed16_over = df['pctemployed16_over'].median()
    df['pctemployed16_over'].fillna(median_pctemployed16_over, inplace=True)

    #median_pctprivatecoveragealone = df['pctprivatecoveragealone'].median()
    df['pctprivatecoveragealone'].fillna(median_pctprivatecoveragealone, inplace=True)

    #median_pctsomecol18_24 = df['pctsomecol18_24'].median()
    df['pctsomecol18_24'].fillna(median_pctsomecol18_24, inplace=True)
    
        # Drop rows with remaining NaN values
    df.dropna(inplace=True)
    
    #Handling errors
        #Percentage
    #pct_percent_columns = [col for col in df.columns if 'pct' in col or 'percent' in col]
    #for col in pct_percent_columns:
     #   df = df[(df[col] > 0) & (df[col] < 100)]
        #medianage
    #if 'medianage' in df.columns:
     #   df = df[df['medianage'] < 100]
    
    # Log transformation
    skew_continuous_data = df.select_dtypes(include=np.number).skew().sort_values()

    high_abs_skew_cols = [col for col in skew_continuous_data.index if abs(skew_continuous_data[col]) >= 1 and col != 'pctwhite']

    # Apply log transformation to columns with skewness above 1 or less than -1, excluding 'pctwhite'
    for column in high_abs_skew_cols:
        df[column] = np.log1p(df[column])
        
    #Outliers removal - zscore
    num_cols = df.select_dtypes(include=np.number).columns
    df[num_cols] = df[num_cols].apply(zscore)   
    
    #One hot encoding
    binnedinc_encoded = pd.get_dummies(df['binnedinc'], prefix='binnedinc')
    df = pd.concat([df.drop(columns=['binnedinc']), binnedinc_encoded], axis=1)
    
    #Label encoding
    label_encoder = LabelEncoder()
    df['geography'] = label_encoder.fit_transform(df['geography'])
    
    return df
```

### (c) Please tell us your detailed machine learning strategy 

First, the data will be split into a training set and a test set. To facilitate hyperparameter tuning, the training set will be further split into training and validation sets. Once EDA is completed, we will define the preprocessing steps to be applied to our various sets. The sets will as well be scaled for ridge regression.

**Pre processing steps :**
The duplicated and erros and NaN will be removed as defined in the EDA part. Outliers with a skewness score above one will be handled with log transformation. The categorical variable about the counties will be label encoded and the remaining categorical variable will be one hot encoded. 

**Models :**

- Baseline Model
    A simple model (e.g., mean predictor) to serve as a reference point.

- Ridge Regression
    Rational: linear regression which handles multicollinearity well and handles overfitting well (regularization)
    
    Tuned Hyperparameters:
    
    *alpha:*  Regularization strength parameter. Higher values mean stronger  regularization. It will prevent overfitting by adding penalty to the loss function.
     alpha values: np.logspace(-4, 4, 50)

- Decision Tree
    Rationale: Robust to noise and can handle missing values (NaNs).
    
    Tuned Hyperparameters:
    
    *Max Depth:* Controls the maximum depth of the tree. A deeper tree can capture more complex patterns but risks overfitting.
    
    *Min Samples Split*: Minimum number of samples required to split an internal node. Helps prevent the tree from creating splits that are based on too few data points.
    
     *Min Samples Leaf*: Minimum number of samples required to be at a leaf node. Ensures that leaf nodes have a certain minimum number of samples.
    
     *Max Features:* The number of features to consider when looking for the best split. This can be set to an exact number, a proportion of the total features, or common choices like the square root or logarithm (base 2) of the total number of features.

- Random Forest
    Tuned Hyperparameters: Same as those for the decision tree, as each tree in the forest uses these parameters.
    

**Tuning :**

All models will be tuned using a combination of grid search and cross-validation.

Grid Search: Iterates over a predefined grid of hyperparameters and evaluates each combination.

Cross-Validation: A statistical technique to evaluate model performance by splitting the data into training and validation sets multiple times. It ensures robustness by averaging the performance across different splits.

We will use GridSearchCV from scikit-learn, which integrates cross-validation within the grid search process. It will automatically split the data, evaluate all combinations, and select the best model based on the cross-validation results.


**Evaluation Metrics :**

In the context of regression models, we use metrics that measure the difference between predicted and actual values:

- Mean Absolute Error (MAE): Measures the average absolute differences between predicted and actual values. Less sensitive to outliers.
    Interpretation: Indicates how wrong predictions are on average. The lower, the better.
    
- Mean Squared Error (MSE): Measures the average of the squared differences between predicted and actual values, penalizing larger errors more.
    Interpretation: Provides an absolute measure of the model’s error magnitude. The lower, the better. Sensitive to outliers.
    
- Root Mean Squared Error (RMSE): The square root of the MSE, providing error magnitude in the same unit as the target variable.
     Interpretation: Useful when penalizing larger errors more. The lower, the better.

- R-squared (R²): Indicates how well the model explains the variance in the target variable. Values range from 0 to 1, with higher values indicating better performance.
     Interpretation: Measures the proportion of variance explained by the model. Higher is better.

Summary
R²: Measures explained variance.
MAE: Treats all errors equally, less sensitive to outliers.
MSE/RMSE: Penalizes larger errors more, influenced by outliers.
